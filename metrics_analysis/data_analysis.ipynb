## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

## Read the Data

```python
platform = 'ec2'

# Load the CSV data into a Pandas DataFrame
df = pd.read_csv(f'data/time_metrics_{platform}_batch2.csv')

# Convert the 'timestamp' column to datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

## Set up the Figures and Plot the Metics

```python
# Create a figure with subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
fig.suptitle('Analysis of Server Data', fontsize=16)

# Plot 1: Requests Handled over Time
sns.lineplot(x='timestamp', y='requests_handled', data=df, ax=axes[0, 0])


# Plot 2: Average CPU Utilization over Time
sns.lineplot(x='timestamp', y='avg_cpu_utilized', data=df, ax=axes[0, 1])


# Plot 3: Words Processed over Time
sns.lineplot(x='timestamp', y='words_processed', data=df, ax=axes[1, 0])


# Plot 4: Success Rate over Time
sns.lineplot(x='timestamp', y='success_rate', data=df, ax=axes[1, 1])


# Plot 5: Average Virtual Memory Utilization over Time
sns.lineplot(x='timestamp', y='avg_vmemory_utilized', data=df, ax=axes[2, 0])


# Plot 6: Number of Paragraphs over Time
sns.lineplot(x='timestamp', y='num_paragraphs', data=df, ax=axes[2, 1])


# Adjust layout and show plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(f'graphs/metrics_{platform}_batch2')
```
